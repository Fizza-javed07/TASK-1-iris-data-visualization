### Introduction and Problem Statement

The Iris dataset is a well-known dataset in data science and machine learning. It contains measurements of three species of iris flowers: Setosa, Versicolor, and Virginica. Each sample includes four features — sepal length, sepal width, petal length, and petal width — along with the species label.

The objective of this task is to explore and visualize the Iris dataset using Python libraries such as pandas, matplotlib, and seaborn. Through this task, we aim to understand the structure of the dataset, examine relationships between features, and identify patterns that could be useful for classification models in future tasks.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

## Dataset Overview

The Iris dataset contains 150 rows and 5 columns. Each row represents a flower sample, and the columns include four numerical features — sepal length, sepal width, petal length, and petal width — along with the species label. All feature columns are of type `float64`, and the species column is of type `object`, indicating categorical data.


```python
df.describe()
```

## Descriptive Statistics Summary

The descriptive statistics show that petal length and petal width have a wider range and higher variability compared to sepal measurements. This suggests that petal features may be more useful for distinguishing between iris species. The mean and standard deviation values also help us understand the typical size and spread of each feature.


```python
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species")
plt.title("Petal Length vs Petal Width by Species")
plt.show()
```

## Scatter Plot Analysis

The scatter plot shows a clear separation between iris species based on petal length and petal width. Setosa is distinctly clustered in the lower-left corner, while Versicolor and Virginica are more spread out but still form separate groups. This suggests that petal measurements are effective features for classifying iris species.


```python
sns.pairplot(df, hue="species")
plt.suptitle("Pairplot of Iris Features by Species", y=1.02)
plt.show()
```

## Pairplot Analysis

The pairplot shows clear patterns and separations between the three iris species. Setosa is distinctly separated from the other two species across all feature combinations. Versicolor and Virginica show some overlap, but there are still visible trends that help differentiate them. This visualization confirms that petal measurements are especially useful for classification.


```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, orient="h")
plt.title("Boxplot of Iris Features")
plt.show()
```

## Boxplot Analysis

The boxplot shows the distribution of each feature across all samples. Petal length and petal width have wider ranges and more variation compared to sepal features. There are also some outliers, especially in sepal width. This visualization helps identify which features are more consistent and which vary more across species.


## Conclusion

The Iris dataset provides a clear example of how numerical features can be used to classify species. Through descriptive statistics and visualizations like scatter plots, pairplots, and boxplots, we observed that petal measurements are especially effective for distinguishing between Setosa, Versicolor, and Virginica. This analysis highlights the power of exploratory data analysis in understanding patterns and preparing for machine learning tasks.
